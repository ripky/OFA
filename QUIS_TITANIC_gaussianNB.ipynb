```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
# panggil data set 
data = pd.read_csv("Titanic (1).csv")

# mengkonversi variabel kategori menjadi data numerik pada variabel sex
data["Sex_cleaned"]=np.where(data["Sex"]=="male",0,1)
```

```python
data
```

```python
#melihat data yang kosong
data.isna().sum()
```

```python
#menghapus data yang kosong yang sudah terlihat pada kode diatas 
data.dropna(inplace=True)
data
```

```python
# menghapus beberapa variabel dan menyisakan variabel yang diinginkan
data=data[[
    "Survived",
    "Sex_cleaned",
    "Age"
]].dropna(axis=0, how='any')
```

```python
data
```

```python
#noralisasi pada variabel jika nilai lebih dari 1
from sklearn.preprocessing import MinMaxScaler
#nilai normalisasi 0 sampai 1
scaler = MinMaxScaler()
#variabel yang ingin di normalkan 
normal=['Age']
data[normal]=scaler.fit_transform(data[normal])
```

```python
data
```

```python
#membuat variabel baru dengan mengisi variabel pada dataset yang ingin diolah
used_features =[
    "Sex_cleaned",
    "Age"
]
```

```python
#membuat variabel dependen dan independen 
x = data[used_features].values
y = data["Survived"]
```

```python
#split dataset menjadi 2, dan membagi lagi menjadi 4(trainX,testX,trainY,testY)
from sklearn.model_selection import train_test_split
trainX, testX, trainY, testY= train_test_split(x, y, test_size=0.2, random_state=15, stratify=y)
```

```python
#untuk melihat berapa banyak jumlah pada variabel data split
print(f'train x:{len(trainX)}')
print(f'train y:{len(trainY)}')
print(f'test x:{len(testX)}')
print(f'test y:{len(testY)}')
```

**train result**

```python
#masukkan kode untuk fungsi gaussian 
gnb=GaussianNB
gnb = GaussianNB()
#masukkan data split train pada kode model gaussian
gnb.fit(trainX,trainY)
#prediksi untuk data train
y_pred_train=gnb.predict(trainX)
```

```python
#import aacuracy 
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
#print dan jalankan fungsi dari accuracy
print("Classification Report is:\n",classification_report(trainY,y_pred_train))
print("Confusion Matrix:\n",confusion_matrix(trainY,y_pred_train))
print("Training Score:\n",gnb.score(trainX,trainY)*100)
```

**test result**

```python
#masukkan kode untuk fungsi gaussian 
gnb=GaussianNB
gnb = GaussianNB()
#masukkan data split train pada kode model gaussian
gnb.fit(testX,testY)
#prediksi untuk data train
y_pred_test=gnb.predict(testX)
```

```python
#print dan jalankan fungsi dari accuracy
print("Classification Report is:\n",classification_report(testY,y_pred_test))
print("Confusion Matrix:\n",confusion_matrix(testY,y_pred_test))
print("Training Score:\n",gnb.score(testX,testY)*100)
```
